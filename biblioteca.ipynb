```python
# Lista de faixas e categorias
faixas_categorias = [
    (0, 99, "Generalidades. Ciência e conhecimento."),
    (100, 199, "Filosofia e psicologia."),
    (200, 299, "Religião."),
    (300, 399, "Ciências sociais."),
    (400, 499, "Classe vaga. Provisoriamente não ocupada."),
    (500, 599, "Matemática e ciências naturais."),
    (600, 699, "Ciências aplicadas."),
    (700, 799, "Belas artes."),
    (800, 899, "Linguagem. Língua. Linguística."),
    (900, 999, "Geografia. Biografia. História.")
]

# Inicializando a nova coluna
categorias = []

# Loop para categorizar
for valor in df_unificado['localizacao']:
    categoria_atual = "Fora da faixa definida."
    for faixa in faixas_categorias:
        if faixa[0] <= valor <= faixa[1]:
            categoria_atual = faixa[2]
            break
    categorias.append(categoria_atual)

# Adicionando a nova coluna ao DataFrame
df_unificado['categoria'] = categorias

df_unificado.sample()
```

```python
df_unificado = df_unificado.drop(columns=['registro_sistema'])
df_unificado
```

```python
emprestimos_data = pd.DataFrame(df_unificado['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns = ['data','quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data
```

```python
emprestimos_por_ano = emprestimos_data.groupby(by=emprestimos_data['data'].dt.year)['quantidade'].sum().reset_index()
emprestimos_por_ano.columns = ['ano', 'total']
emprestimos_por_ano
```

```python
sns.set_theme(context='notebook', 
              style='darkgrid', 
              palette='deep', 
              font_scale=1.3, 
              rc={"figure.figsize":(15,8)})
```

```python
from matplotlib import ticker

ax = sns.lineplot(data=emprestimos_por_ano,x='ano',y='total')
ax.set(xlabel=None,ylabel=None)
ax.tick_params(axis='x', rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))   

ax.set_title('Quantidade de exemplares emprestados do SISBI por ano'+'\n',size=20,loc='left',weight='bold')

ax=ax
```

```python
emprestimos_por_mes = emprestimos_data.groupby(by=emprestimos_data['data'].dt.month)['quantidade'].sum().reset_index()
emprestimos_por_mes
```

```python
dicionario_meses = {1:'Jan',2:'Fev',3:'Mar',4:'Abr',
                    5:'Mai',6:'Jun',7:'Jul',8:'Ago',
                    9:'Set',10:'Out',11:'Nov',12:'Dez'}
```

```python
emprestimos_por_mes.columns = ['mes', 'total']
emprestimos_por_mes['mes'] = emprestimos_por_mes['mes'].map(dicionario_meses)
emprestimos_por_mes
```

```python
ax = sns.lineplot(data=emprestimos_por_mes,x='mes',y='total')
ax.set(xlabel=None,ylabel=None)
ax.tick_params(axis='x',rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))                

ax.set_title("Quantidade de exemplares emprestados do SISBI por mês"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=265000,fontsize=18, ha='left',color='gray')

ax=ax
```

```python
emprestimos_por_hora = emprestimos_data.groupby(by=emprestimos_data['data'].dt.hour)['quantidade'].sum().reset_index()
emprestimos_por_hora.columns = ['hora', 'total']
emprestimos_por_hora
```

```python
emprestimos_por_hora = emprestimos_por_hora.sort_values(ascending=True,by='total')

ax = sns.barplot(data=emprestimos_por_hora,y='total',x='hora',
                 palette='Blues', hue='total',dodge=False)                
plt.legend([],[], frameon=False)                                              

ax.set(xlabel='Horário',ylabel=None)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))           
ax.set_title("Quantidade de exemplares emprestados do SISBI por faixa horária"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Período de 2010 a 2020',x=-0.5,y=225000,fontsize=18, ha='left',color='gray')
ax=ax
```
